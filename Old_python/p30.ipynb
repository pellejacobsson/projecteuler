```python
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
def psum(n, p):
    m = [int(s)**p for s in str(n)]
    return sum(m)
```

```python
psum(10000, 4)
```

```python
n = range(2, 1000000)
ns = []
for nn in n:
    ns.append(psum(nn, 5))
```

```python
plt.figure()
plt.plot(n, ns, '.')
plt.plot([0, n[-1]], [0, n[-1]])
```

```python
s = 0
for n in range(2, 300002):
    np = psum(n, 5)
    if n == np:
        print(n)
        s += n
    
```
