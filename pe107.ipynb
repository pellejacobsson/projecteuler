```python
import networkx as nx
import matplotlib as plt
```

```python
with open("0107_network.txt") as f:
    d = {}
    for (i, line) in enumerate(f):
        for (j, s) in enumerate(line.strip().split(",")):
            if s != "-":
                d[i, j] = int(s)
```

```python
g = nx.Graph()
```

```python
for (i, j), weight in d.items():
    g.add_edge(i, j, weight=weight)
```

```python
mst = nx.minimum_spanning_tree(g)
```

```python
nx.draw(mst)
```

```python
original_weight = sum([e[2]["weight"] for e in g.edges(data=True)])
```

```python
minimum_weight = sum([e[2]["weight"] for e in mst.edges(data=True)])
```
